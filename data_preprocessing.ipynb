### Clean Text 

```python
#StopWords

stopWords = [
    "ပါ", "က", "တယ်", "ကို", "နေ", "တာ", "သည်", "မှာ", "တွေ", "ရ", 
    "တော့", "ဖြစ်", "တဲ့", "များ", "ပြီး", "လည်း", "နဲ့", "လို", "တို", "ရှိ", 
    "ဆို", "ပဲ", "သူ", "ခဲ့", "ဘူး", "တစ်", "လိုက်", "မှ", "ကြ", "လာ", "သွား", 
    "ပြော", "ပေး", "၏", "လား", "ရင်", "နိုင်", "ထား", "သော", "လေ", "မယ်", 
    "လေး", "ပြီ", "ချင်", "လုပ်", "တွင်", "လဲ", "နှင့်", "ဟုတ်", 
     "ရဲ့", "နော်", "ဘာ", "ရေး", "ဒီ", "မှု", "လဲ", "ဟာ", 
    "အရမ်း", "ပြန်", "နှစ်", "ပါစေ", "ဖို့", "ပြီ", "လား", "ခု", "ဖို့", "ရေ", "ဦး", "အောင်", "ရောက်", "ဘာ", "ဟု", 
    "နောက်", "ကဲ", "မူ", "အတွက်", "ကြီး", "ထဲ", "ခြင်း", "စေ", "သေး", "မည်","မိ",
    "၍","ဟယ်","ရှင့်","ဟင်",
]
```

```python
import re
import ast

# Function to remove emojis, special characters, and clean text
def remove_emoji_and_special_char(text):
   #Remove Invisible Unicode Characters
    text = re.sub(r'[\u200B\u200C\u200D\u2060\uFEFF]', '', text)
    
    #Remove Corrupted or Placeholder Unicode
    text = re.sub(r'[\uFFFD\u25A1\u25A0]', '', text)

    #Keeps English alphabets (a-z, A-Z),Burmese Unicode block (\u1000-\u109F),Common Burmese diacritics and signs (\u102C, \u103D, etc.),Spaces, brackets, and quotes for formatting
    text = re.sub(r'[^\x00-\x7F\u1000-\u109F\u102C\u102D\u102E\u1036\u1039\u103A\u103C\u103D\u1030\u1031\u1032\u1037\u1038\s\[\]\,\'\"a-zA-Z]', '', text)
    #Remove Control Characters and Formatting Marks
    text = re.sub(r'[\x00-\x1F\x7F-\x9F\u202A-\u202E]', '', text)



    #Add Separator Between Burmese and English Words
    #(e.g., helloမင်္ဂလာပါ → hello,မင်္ဂလာပါ)
    text = re.sub(r'([a-zA-Z]+)([က-အ])', r'\1,\2', text)
    text = re.sub(r'([က-အ])([a-zA-Z]+)', r'\1,\2', text)

    #Strip Punctuation
    text = text.replace('"', '').replace("", "")  # Remove all double quotes
    text = re.sub(r'[:;!?.\\\(\)]', '', text)  # Remove common punctuation

    #Normalize Multiple Whitespace to Single Whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    emoji_pattern = re.compile(
        "[" +
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed characters
        "]+", flags=re.UNICODE
    )
    #Remove Emojis
    text = emoji_pattern.sub('', text)

    return text



```

```python
# StopWord Removal
df["clean_text"] = df["tokenized_comments"].apply(
    lambda x: [
        word for cleaned_word in [remove_emoji_and_special_char(word) for word in x if isinstance(word, str)]
        for word in cleaned_word.split(',') # Split concatenated words (e.g., EnglishBurmese)
        if word not in stopWords and word.strip() != ''
    ]
)


df[["clean_text"]].head(20)
```
